```python
import numpy as np
import matplotlib.pyplot as plt

# Definir parmetros
fs = 100  # Frecuencia de muestreo
T = 1     # Duracin de la seal en segundos
t = np.linspace(0, T, fs, endpoint=False)  # Vector de tiempo

# Seales sinusoidales
f1 = 5  # Frecuencia de la primera seal
f2 = 10 # Frecuencia de la segunda seal

# Crear la primera seal como una sinusoide truncada
x = np.sin(2 * np.pi * f1 * t)
x[fs//2:] = 0  # Hacer que la segunda mitad sea 0

# Segunda seal como sinusoide completa
h = np.sin(2 * np.pi * f2 * t)

# Calcular la convolucin
y = np.convolve(x, h, mode='full')
n = np.arange(len(y))  # Eje de tiempo para la convolucin

# Seleccionar la parte de la convolucin a graficar
start_idx = 0     # ndice inicial
end_idx = 200     # ndice final

n_partial = n[start_idx:end_idx]
y_partial = y[start_idx:end_idx]

# Graficar la seal de entrada y la convolucin parcial
plt.figure(figsize=(10, 4))

plt.subplot(2, 1, 1)
plt.plot(t, x, label="Seal de entrada x[n]")
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.title('Seal de Entrada (Sinusoide Truncada)')
plt.legend()
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(n_partial, y_partial, label="Parte de la convolucin")
plt.xlabel('n')
plt.ylabel('y[n]')
plt.title('Parte de la Convolucin que luego se hace 0')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()

```
