```python
import numpy as np
import matplotlib.pyplot as plt

# Definir parámetros
fs = 100  # Frecuencia de muestreo
T = 1     # Duración de la señal en segundos
t = np.linspace(0, T, fs, endpoint=False)  # Vector de tiempo

# Señales sinusoidales
f1 = 5  # Frecuencia de la primera señal
f2 = 10 # Frecuencia de la segunda señal

# Crear la primera señal como una sinusoide truncada
x = np.sin(2 * np.pi * f1 * t)
x[fs//2:] = 0  # Hacer que la segunda mitad sea 0

# Segunda señal como sinusoide completa
h = np.sin(2 * np.pi * f2 * t)

# Calcular la convolución
y = np.convolve(x, h, mode='full')
n = np.arange(len(y))  # Eje de tiempo para la convolución

# Seleccionar la parte de la convolución a graficar
start_idx = 0     # Índice inicial
end_idx = 200     # Índice final

n_partial = n[start_idx:end_idx]
y_partial = y[start_idx:end_idx]

# Graficar la señal de entrada y la convolución parcial
plt.figure(figsize=(10, 4))

plt.subplot(2, 1, 1)
plt.plot(t, x, label="Señal de entrada x[n]")
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.title('Señal de Entrada (Sinusoide Truncada)')
plt.legend()
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(n_partial, y_partial, label="Parte de la convolución")
plt.xlabel('n')
plt.ylabel('y[n]')
plt.title('Parte de la Convolución que luego se hace 0')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()

```
